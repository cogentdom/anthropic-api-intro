## Chat

#### Setup

```python
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"

# Loads variables from the environment
from os import getenv
from dotenv import load_dotenv
load_dotenv()
open_api_key = getenv("ANTHROPIC_API_KEY")
```

#### Helper Functions

```python
def add_user_message(messages, text):
    user_message = {
        "role": "user",
        "content": text
    }
    messages.append(user_message)
    
def add_assistant_message(messages, text):
    assistant_message = {
        "role": "assistant",
        "content": text
    }
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages
    )
    return message.content[0].text
```

#### Exercise

```python
# Make an initial list of messagaes
messages = []

# use a While true loop to run the chatbot forever
while True:
    # get user input
    user_input = input("> ")
    print(">",user_input)
    
    # Add the user input to the messages list
    add_user_message(messages,user_input)
    # Call Claude with the 'chat' funciton
    answer = chat(messages)
    # Add generated text to the liust of messages 
    add_assistant_message(messages, answer)
    # Print the generated text
    print("---")
    print(answer)
    print("---")
    print()
```
